```python
import pandas as pd
import numpy as np
import seaborn as sns

import matplotlib.pyplot as plt
import matplotlib.mlab as mlab
import matplotlib
plt.style.use('ggplot')
from matplotlib.pyplot import figure

%matplotlib inline
matplotlib.rcParams['figure.figsize'] = (12,8)

#Read the data.

df = pd.read_csv(r'C:\Users\SHIVAKUMAR S B\Downloads\movies.csv')
```

```python
# Lets's check at the data
df.head()
```

```python
# Checking to see if we have any missing data
# Let's loop through the data and see if there is anything missing

for col in df.columns:
    pct_missing = np.mean(df[col].isnull())
    print('{} - {}%'.format(col, round(pct_missing*100)))
    
    df = df.dropna()
```

```python
#lets check Data Types for our columns

print(df.dtypes)
```

```python
#change data type of colunms

df['budget']= df['budget'].astype('int64')
 
df['gross'] = df['gross'].astype('int64')
```

```python
df
```

```python
#Create correct year column.
df['yearcorrect'] = df['released'].str.extract(pat = '([0-9]{4})').astype(int)
df
```

```python
#sorting data for highest grossed movie in ascending order.
df=df.sort_values(by=['gross'], inplace=False,ascending=False)
```

```python
df.head()
```

```python
#points to know which have high possiblities of correlation
#budget high correlation
#company high correlation
```

```python
# scatter plot with budget vs gross

plt.scatter(x=df['budget'],y=df['gross'])

plt.title('budget vs gross earnings')

plt.xlabel('budget of movies')

plt.ylabel('gross earnings')

plt.show()
```

```python
#ploting budget vs gross using seaborn
sns.regplot(x='budget', y='gross', data=df,scatter_kws={"color":"red"}, line_kws={"color":"black"})
```

```python
# lets start looking at correlation
```

```python
df.corr(method='pearson') #pearson,kendall,spearman
```

```python
#found out high correlation is between budget and gross.
```

```python
#Correlation matrix for Numeric Features
correlation_matrix= df.corr(method='pearson')

sns.heatmap(correlation_matrix, annot=True)

plt.title('Correlation matrix for Numeric Features')

plt.xlabel('Movie Features')

plt.ylabel('Movie Features')

plt.show
```

```python
#Looks at companies now
df_numerized = df


for col_name in df_numerized.columns:
    if(df_numerized[col_name].dtype == 'object'):
        df_numerized[col_name]= df_numerized[col_name].astype('category')
        df_numerized[col_name] = df_numerized[col_name].cat.codes
        
df_numerized
```

```python
correlation_matrix= df_numerized.corr(method='pearson')

sns.heatmap(correlation_matrix, annot=True)

plt.title('Correlation matrix for Numeric Features')

plt.xlabel('Movie Features')

plt.ylabel('Movie Features')

plt.show
```

```python
df_numerized.corr()
```

```python
#unstacking the data
Correlation_mat=df_numerized.corr()

Corr_pairs=Correlation_mat.unstack()

Corr_pairs
```

```python
#sorting the pairs from unstacking the data
Sorted_pairs=Corr_pairs.sort_values()

Sorted_pairs
```

```python
# Looking at the top 15 compaies by gross revenue

CompanyGrossSum = df.groupby('company')[["gross"]].sum()

CompanyGrossSumSorted = CompanyGrossSum.sort_values('gross', ascending = False)[:15]

CompanyGrossSumSorted = CompanyGrossSumSorted['gross'].astype('int64') 

CompanyGrossSumSorted
```
